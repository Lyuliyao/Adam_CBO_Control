```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
from matplotlib.ticker import FuncFormatter
from matplotlib import rc

# pwd()
rc("text",usetex="True")
rc("font", family="serif")
# Enable LaTeX rendering
# rc('text', usetex=True)
```

```python
colormap = mpl.colormaps["tab20"]
# colormap get the first color
color = colormap(0)
color
# Function to format scientific notation in LaTeX-like style
def format_sci(value):
    base, exponent = f"{value:.1e}".split("e")
    return f"${base} \\times 10^{{{int(exponent)}}}$"
```

```python

dim = [1, 2, 4, 8, 16]
cbo_value_silu = []
cbo_value_silu_mcbo = []

han_2018_value = []
y_start = []

# Read the data
for d in dim:
    data = pd.read_csv(f"../LQ/BSDE/case_case1_hjb_dim_{d:03d}.csv")
    y_start.append(data["yExact"].values[-1])
    han_2018_value.append(data["yMean"].values[-1])

for d in dim:
    data = np.load(f"../LQ/Adam_CBO_case1/result_{d}/result.npz")
    cbo_value_silu.append(data["cbo_value"][0])

for d in dim:
    data = np.load(f"../LQ/M_CBO_case1/result_{d}/result.npz")
    cbo_value_silu_mcbo.append(data["cbo_value"][0])


# Plotting
plt.figure(figsize=(6, 6))
plt.subplots_adjust(left=0.15, right=0.95, top=0.95, bottom=0.15)
plt.semilogx(dim, han_2018_value, "-.", label="BSDE", color=colormap(0), linewidth=2)
plt.semilogx(dim, cbo_value_silu_mcbo, "-*", label="M-CBO", color=colormap(6), linewidth=2,markersize=4)
plt.semilogx(dim, cbo_value_silu, "-<", label="Adam-CBO", color=colormap(4), linewidth=2)
plt.semilogx(dim, y_start, label="MC Estimation", color=colormap(2), linewidth=2)

# Adding text with formatted values
for i, d in enumerate(dim):
    han_diff = abs(han_2018_value[i] - y_start[i]) / y_start[i]
    cbo_diff = abs(cbo_value_silu[i] - y_start[i]) / y_start[i]
    mcbo_diff = abs(cbo_value_silu_mcbo[i] - y_start[i]) / y_start[i]
    plt.text(d / 2.2, han_2018_value[i] + 0.15, format_sci(han_diff), fontsize=20, color=colormap(0))
    plt.text(d / 0.85, cbo_value_silu[i] - 0.12, format_sci(cbo_diff), fontsize=20, color=colormap(4))
    plt.text(d / 0.85, cbo_value_silu[i] + 0.1, format_sci(mcbo_diff), fontsize=20, color=colormap(6))

# Labels and ticks
plt.xlabel("Dimension", fontsize=32)
plt.ylabel("Value Function", fontsize=32)
plt.xticks(dim, labels=[r"${}$".format(d)  for d in dim], fontsize=32)
plt.yticks(np.arange(0,5,1),fontsize=32)
plt.legend(fontsize=20)
plt.xlim([0.5, 45])
plt.ylim([-0.3, 4])
plt.tick_params(direction="in", top=True, right=True,length=10)
plt.tick_params(direction="in", top=True, right=True,length=10)
# Save figures
plt.savefig("value_function_case1.eps")
plt.savefig("value_function_case1.pdf")
```

```python

# Load your data
data_han = np.load("../LQ/value_function_case1/data_hjb_1D_plot.npz")
data_est = np.load("../LQ/value_function_case1/data_hjb_1D_plot_est.npz")
data_cbo = np.load("../LQ/value_function_case1/cbo_hjb_1D.npz")

# Create subplots, sharing the y-axis, with no horizontal space between them
fig, axs = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Adjust the layout to remove space between the subplots
fig.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.15, wspace=0)

# Plot the first contour
contour1 = axs[0].contourf(data_han["X"], data_han["t"], data_han["res"],levels=np.arange(-1,2,0.2),cmap='RdBu_r')
axs[0].set_title('BSDE', fontsize=32)
axs[0].set_xticks(np.arange(-2, 3, 1))
axs[0].set_xlim([-2.8,2.8])
axs[0].set_yticks(np.arange(0, 1.1, 0.3))
axs[0].tick_params(axis='both', which='major', labelsize=32)

# Plot the second contour
contour2 = axs[1].contourf(data_han["X"], data_han["t"], data_est["res"],levels=np.arange(-1,2,0.2),cmap='RdBu_r')
axs[1].set_title('MC Estimation', fontsize=32)
axs[1].set_xticks(np.arange(-2, 3, 1))
axs[1].set_xlim([-2.8,2.8])

axs[1].tick_params(axis='x', which='major', labelsize=32)

# Plot the third contour
contour3 = axs[2].contourf(data_han["X"], data_han["t"], data_cbo["cbo_value"][..., 0],levels=np.arange(-1,2,0.2),cmap='RdBu_r')
axs[2].set_title('Adam-CBO', fontsize=32)
axs[2].set_xticks(np.arange(-2, 3, 1))
axs[2].set_xlim([-2.8,2.8])
axs[2].tick_params(axis='x', which='major', labelsize=32)

# Add a single color bar that spans all three plots
cbar = fig.colorbar(contour3, ax=axs, location='right', fraction=0.046, pad=0.04,ticks=np.arange(-1,3,0.8))
cbar.ax.tick_params(labelsize=32)

# Set the y-axis label on the first subplot
fig.supylabel(r'$t$', fontsize=32)
fig.supxlabel(r'$x$', fontsize=32)


axs[0].tick_params(direction="in",top=True, right=True,length=10)
axs[1].tick_params(direction="in",top=True, right=True,length=10)
axs[2].tick_params(direction="in",top=True, right=True,length=10)

plt.savefig("value_function_1D_case1.eps")
plt.savefig("value_function_1D_case1.pdf")
```

```python
import scipy.stats as stats

# Create subplots, sharing the y-axis, with no horizontal space between them
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Adjust the layout to remove space between the subplots
fig.subplots_adjust(left=0.05, right=0.95, top=0.98, bottom=0.08, wspace=0)
x_save = np.load("../control_multi_agent/cbo_agent2/result_4/x_save.npy")
for i in range(100):
    x = x_save[:, i, 0]  # x positions
    y = x_save[:, i, 1]  # y positions
    axs[0].plot(x, y, color=colormap(2), alpha=0.2)  # faded lines for sample trajs
    axs[0].plot(x[0], y[0], 'o', color= colormap(2),  markersize=4)  # start point
    axs[0].plot(x[-1], y[-1], 'o', color= colormap(2) , markersize=4)  # end point

    x = x_save[:, i, 2]  # x positions
    y = x_save[:, i, 3]  # y positions
    axs[0].plot(x, y, color=colormap(4), alpha=0.2)  # faded lines for sample trajs
    axs[0].plot(x[0], y[0], 'o',color= colormap(4), markersize=4)  # start point
    axs[0].plot(x[-1], y[-1], 'o', color= colormap(4), markersize=4)  # end point
x =  np.arange(-1,10,0.01)
y =  np.arange(-3,3,0.01)
radius =1.2
distance =  np.sqrt(((x[None,:]-2.5)**2+y[:,None]**2))
soft_zone = (distance > radius) & (distance <= radius * 1.2)
soft_penalty = soft_zone.astype(np.float64) * (
    0.5 + 0.5 * np.cos(np.pi * (distance - radius) / (radius * 0.2))
)
hard_penalty = (distance <= radius).astype(np.float64)
z = hard_penalty + soft_penalty
axs[0].contourf(x, y , z ,levels=np.arange(0,3.1,0.25),cmap='Blues')

x = [0,0]
y = [1,-1]
axs[0].plot(x,y,"*", color= colormap(6),markersize=20,label="start")

x = [5,5]
y = [1,-1]
axs[0].plot(x,y,"*", color= colormap(8),markersize=20,label="target")


# axs[0].contourf(data["X"][1:-1,1:-1],data["Y"][1:-1,1:-1],data["action"][1:-1,1:-1,0,0],levels=np.arange(-1,1,0.05),cmap='RdBu_r')
# axs[0].set_title('Our method', fontsize=20)
axs[0].set_xticks(np.arange(0, 6, 2 ))
axs[0].set_yticks(np.arange(-4, 4, 2))
axs[0].set_xlim([-0.5,5.5])
axs[0].set_ylim([-3,3])
axs[0].tick_params(axis='both', which='major', labelsize=32)
axs[0].legend(fontsize=24)



x_save = np.load("../control_multi_agent/cbo_agent4/result_8/x_save.npy")
for i in range(100):
    for k in range(4):
        x = x_save[:, i, 2*k]  # x positions
        y = x_save[:, i, 2*k+1]  # y positions
        axs[1].plot(x, y, color=colormap(2+k), alpha=0.2)  # faded lines for sample trajs
        axs[1].plot(x[0], y[0], 'o', color= colormap(2+k),  markersize=4)  # start point
        axs[1].plot(x[-1], y[-1], 'o', color= colormap(2+k) , markersize=4)  # end point



r = 1.0

obstacles_one_petal = [
    ((2.0, 0.0), r),
    ((2.0 + np.sqrt(3)/2 * r, 0.5 * r), r),
    ((2.0 + np.sqrt(3)/2 * r, -0.5 * r), r),
]

def rotate_point(p, angle, origin=(0, 0)):
    """Rotate point p around origin by angle (radians)."""
    ox, oy = origin
    px, py = p
    qx = ox + np.cos(angle) * (px - ox) - np.sin(angle) * (py - oy)
    qy = oy + np.sin(angle) * (px - ox) + np.cos(angle) * (py - oy)
    return (qx, qy)

center = (7.5, 0.0)

obstacles = []

for i in range(5):
    angle = i * 2 * np.pi / 5 
    for (pos, radius) in obstacles_one_petal:
        rotated_pos = rotate_point(pos, angle, origin=center)
        obstacles.append((rotated_pos, radius))


def smooth_penalty(distance: np.ndarray, radius: float) -> np.ndarray:
    soft_zone = (distance > radius) & (distance <= radius * 1.2)
    hard_penalty = (distance <= radius).astype(np.float64)
    soft_penalty = soft_zone.astype(np.float64) * (
        0.5 + 0.5 * np.cos(np.pi * (distance - radius) / (radius * 0.2))
    )
    return hard_penalty + soft_penalty

x =  np.arange(-2,20,0.01)
y =  np.arange(-10,10,0.01)
z = 0*x[None,:]*y[:,None]
for pos, radius in obstacles:
    dist = np.sqrt((x[None,:] - pos[0])**2+(y[:,None] - pos[1])**2)
    z = z +  smooth_penalty(dist,radius)
dist = np.sqrt((x[None,:] - 7.5)**2+(y[:,None] - 0.0)**2)
z = z +  smooth_penalty(dist,1.5*radius)

axs[1].contourf(x, y , z ,levels=np.arange(0,3.1,0.25),cmap='Blues')

x = [0,0,0,0]
y = [3,1,-1,-3]
axs[1].plot(x,y,"*", color= colormap(6),markersize=20,label="start")

x = [18,16,16,18]
y = [2,2,-2,-2]
axs[1].plot(x,y,"*", color= colormap(8),markersize=20,label="target")
axs[1].set_xticks(np.arange(0, 20, 4 ))
axs[1].set_yticks(np.arange(-6, 9, 4))
axs[1].set_xlim([-1.5,18.5])
axs[1].set_ylim([-10,10])
axs[1].tick_params(axis='both', which='major', labelsize=32)
axs[1].legend(fontsize=24)

import matplotlib.cm as cm
cmap = cm.get_cmap('tab20', 50)
x_save = np.load("../control_multi_agent/cbo_agent50/result_100/x_save.npy")
for i in range(100):
    for k in range(50):
        x = x_save[:, i, 2*k]  # x positions
        y = x_save[:, i, 2*k+1]  # y positions
        axs[2].plot(x, y, color=cmap(k), alpha=0.2)  # faded lines for sample trajs
        axs[2].plot(x[0], y[0], 'o', color= cmap(k),  markersize=4)  # start point
        axs[2].plot(x[-1], y[-1], 'o', color= cmap(k) , markersize=4)  # end point


        
obstacles = [
    (np.array([9.0, 15.0])/10, 2/10),
    (np.array([8.0, 9.0])/10, 1.8/10),
    (np.array([7.0, 3.0])/10, 2.0/10),
    (np.array([9.0, -2.0])/10, 1.2/10),
    (np.array([12.0, -6.0])/10, 1.7/10),
    (np.array([10.0, -13.0])/10, 1.5/10),
]


def hard_penalty(distance: np.ndarray, radius: float) -> np.ndarray:
    hard_penalty = (distance <= radius).astype(np.float64)
    return hard_penalty


x =  np.arange(-2,6,0.01)
y =  np.arange(-5,5,0.01)
z = 0*x[None,:]*y[:,None]
for pos, radius in obstacles:
    dist = np.sqrt((x[None,:] - pos[0])**2+(y[:,None] - pos[1])**2)
    z = z +  hard_penalty(dist,radius)

contour1=axs[2].contourf(x, y , z ,levels=np.arange(0,3.1,0.25),cmap='Blues')


x_start = np.zeros((100))
for i in range(10):
    for j in range(5):
        x_start[10*i+2*j]  = (-10/3*j)
        x_start[10*i+2*j+1]= (10/3*i-15)
x_start = x_start/10
axs[2].plot(x_start[::2],x_start[1::2],"*", color= colormap(6),markersize=10,label="start")


x_target = np.zeros((100))
for i in range(10):
    for j in range(5):
        x_target[10*i+2*j]   =(20+10/2*j)
        x_target[10*i+2*j+1] =(10/4*i-45/4)
x_target = x_target/10
axs[2].plot(x_target[::2],x_target[1::2],"*", color= colormap(8),markersize=10,label="target")
axs[2].set_xticks(np.arange(-0, 5, 2 ))
axs[2].set_yticks(np.arange(-2, 4, 2))
axs[2].set_xlim([-1.5,4.5])
axs[2].set_ylim([-3,3])
axs[2].tick_params(axis='both', which='major', labelsize=32)
axs[2].yaxis.tick_right()
axs[2].legend(fontsize=24)
cbar = fig.colorbar(contour1, ax=axs, location='right', fraction=0.046, pad=0.04,ticks=np.arange(0,3.1,1))

cbar.ax.tick_params(labelsize=32)

plt.savefig("path.pdf")
plt.savefig("path.png")
```
